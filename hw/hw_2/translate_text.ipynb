```python
from transformers import MarianMTModel, MarianTokenizer

model_name_from_en_to_ru = "Helsinki-NLP/opus-mt-en-ru"
tokenizer_from_en_to_ru = MarianTokenizer.from_pretrained(model_name_from_en_to_ru)
model_from_en_to_ru = MarianMTModel.from_pretrained(model_name_from_en_to_ru)

model_name_from_ru_to_en = "Helsinki-NLP/opus-mt-ru-en"
tokenizer_from_ru_to_en = MarianTokenizer.from_pretrained(model_name_from_ru_to_en)
model_from_ru_to_en = MarianMTModel.from_pretrained(model_name_from_ru_to_en)


def translate_from_ru_to_en(text):
    batch = tokenizer_from_ru_to_en([text], return_tensors="pt", padding=True)
    generated_ids = model_from_ru_to_en.generate(**batch)
    return tokenizer_from_ru_to_en.batch_decode(
        generated_ids, skip_special_tokens=True
    )[0]


def translate_from_en_to_ru(text):
    batch = tokenizer_from_en_to_ru([text], return_tensors="pt", padding=True)
    generated_ids = model_from_en_to_ru.generate(**batch)
    return tokenizer_from_en_to_ru.batch_decode(
        generated_ids, skip_special_tokens=True
    )[0]


result = translate_from_en_to_ru("Hello world")
print(result)
```

```python
import pandas as pd

df = pd.read_csv("./data/train.csv", nrows=1_000)
```

```python
import re


original_text = df[df["rating"] == 3]["text"].iloc[2]


def clean_text(text):
    # Убираем лишние обратные слеши и пробелы
    text = text.replace("\\n", " ")

    # Убираем лишние пробелы
    text = re.sub(r"\s+", " ", text)

    # Убираем пробелы перед точками и запятыми
    text = re.sub(r"\s+\.", ".", text)
    text = re.sub(r"\s+,", ",", text)

    return text.strip()


cleaned_text = clean_text(original_text)
print(cleaned_text)
```

```python
translated_text = translate_from_ru_to_en(cleaned_text)
new_text = translate_from_en_to_ru(translated_text)
translated_text = translate_from_ru_to_en(new_text)
new_text = translate_from_en_to_ru(translated_text)
```
